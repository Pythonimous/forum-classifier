```python
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
from bs4 import BeautifulSoup
import requests

from tqdm import tqdm_notebook as tqdm
import time

from collections import defaultdict

import pickle
```

```python
s = requests.Session()
```

```python
with open('threads.txt', 'r') as t:
    pairs = [a.split() for a in t.read().split('\n') if a]
t.close()
```

```python
categories = set([p[1] for p in pairs])
```

```python
url_dict = {c:[] for c in categories}
```

```python
def get_urls_list(url, category_name):
    resp = s.get(url, verify = False)
    soup = BeautifulSoup(resp.text, 'html.parser')
    pagenav = soup.find('div', class_='pagenav')
    last_page_num = int(pagenav.find(nowrap='nowrap').find('a').get('href').split('=')[-1])
    urls_list = [url + f'&page={i}' for i in range(1, last_page_num+1)]
    url_dict[category_name] += urls_list
```

```python
for p in pairs:
    get_urls_list(p[0], p[1])
```

```python
url_dict
```

```python
with open('links.pkl', 'wb') as l:
    pickle.dump(url_dict, l)
```
