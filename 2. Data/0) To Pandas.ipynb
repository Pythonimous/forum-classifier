```python
import pandas as pd
```

```python
def unify(typ):
    
    with open(typ+'_corpus.txt', 'r', encoding='utf-8') as t1:
        corp = t1.read().split('\nNEWTEXT\n')
    t1.close()
    
    corp = [' '.join(a.replace('\n', ' ').split()) for a in corp]
    
    with open(typ+'_labels.txt', 'r', encoding='utf-8') as t2:
        labels = t2.read().split('\n')
    t2.close()
    
    pairs = list(zip(labels, corp))
    df = pd.DataFrame(pairs, columns = ['label', 'text'])
    
    df.to_csv(typ + '.csv')
    
    return df
```

```python
unify('Train')
```

```python
unify('Test')
```

```python
unify('Val')
```
