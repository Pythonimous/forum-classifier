```python
import pickle
```

```python
with open('top_threehundred_ppmi.pkl', 'rb') as hr:
    top = pickle.load(hr)
hr.close()
```

```python
top
```

```python
categs = ['anime', 'art', 'books', 'films', 'food', 'football', 'games', 'music', 'nature', 'travel']
```

```python
allwords = list(set(sum(list(top.values()), [])))
```

```python
occurrences = {} # словарь слово - список порядка вхождений
for w in allwords:
    perdict = [] # список: десять индексов, в которых есть слово
    for c in categs:
        try:
            num = top[c].index(w)
        except ValueError:
            num = 1000
        perdict.append(num)
    occurrences[w] = perdict
```

```python
wheretokeep = {k : val.index(min(val)) for k, val in occurrences.items()}
```

```python
for k, val in wheretokeep.items():
    deletefrom = [i for i in range(10) if i != val]
    for i in deletefrom:
        try:
            top[categs[i]].remove(k)
        except ValueError:
            pass
```

```python
top_sets = {k:set(val) for k, val in top.items()}
```

```python
categs = ['anime', 'art', 'books', 'films', 'food', 'football', 'games', 'music', 'nature', 'travel']
weights = [len(v) / max([len(v) for v in top_sets.values()]) for v in top_sets.values()]
def words_from_tokens(tokens):
    vec = [0] * 10
    for i in range(len(categs)):
        count = 0
        for t in tokens:
            if t in top_sets[categs[i]]:
                count += 1
        vec[i] = (count / len(tokens))/weights[i]
    return vec
```
